### Question 2
Given 
$q(0) = (0.4, 0.2, -0.1)$, write a program to integrate the CRP differential kinematic equations. Assume the body angular velocity vector of the craft is given through the $B$ frame components as $^B\omega = \begin{pmatrix} \sin(0.1t) \\ 0.01 \\ \cos(0.1t) \end{pmatrix}$ 3 deg/s.

Enter below the CRP norm $q_1^2 + q_2^2 + q_3^2$ at 42 seconds into the simulation.


```python
def p_dot(p,t):
    """
    Differential Kinematic Equation
    """
    omega = np.array([np.sin(0.1*t), 0.01, np.cos(0.1*t)]) * np.deg2rad(3.)
    return am.CRP_omega_DKE(p,omega)

p = odeint(p_dot, np.array([0.4, 0.2, -0.1]), np.arange(0.,42.,0.01))
print(np.linalg.norm(p[-1]))
plt(p)
```
